K1 and K2 identify the mutual inductance between two inductors. k is the coefficient of coupling. These need to be converted to mutual inductance.

Need to solve for M1 and M2:  
$M_1 = k \times {\sqrt{L_1 \times L_2}}$  
$M_2 = k \times {\sqrt{L_3 \times L_4}}$  

```python
# calculate the coupling constant from the mutual inductance
component_values[M1] = component_values[K1]*np.sqrt(component_values[L1] *component_values[L2])
print('mutual inductance, M1 = {:.9f}'.format(component_values[M1]))

component_values[M2] = component_values[K2]*np.sqrt(component_values[L3] *component_values[L4])
print('mutual inductance, M1 = {:.9f}'.format(component_values[M2]))
```
